```python
from datetime import date
from dateutil.relativedelta import relativedelta
from pyspark.sql import Row

# Remove a tabela se já existir
spark.sql("DROP TABLE IF EXISTS lk_ControleProcessos.tab_controle_incremental")

# 1. Define a data de corte: primeiro dia de 5 meses atrás
data_base = date.today() - relativedelta(months=5)
data_corte = data_base.replace(day=1)
data_corte_str = data_corte.strftime('%Y-%m-%d')

# 2. Cria o DataFrame com a nova data
df = spark.createDataFrame([Row(vDataCorte=data_corte_str)])

# 3. Sobrescreve a tabela no Lakehouse
df.write.mode("overwrite").saveAsTable("lk_ControleProcessos.tab_controle_incremental")
```
